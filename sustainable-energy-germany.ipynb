# Slicing the Dataset and using only data about Germany 

```python
#slicing the data and getting only the data about Germany 
df_Germany=df[df["Entity"]=="Germany"]
df_Germany
```

```python
#checking if all the data typs in the correct form 
df_Germany.info()
```

```python
#collecting columes names to remove the uneeded columes 
df_Germany.columns
```

```python
#removing uneeded columes for better analysis 
Removed_Columes=["Access to electricity (% of population)","Access to clean fuels for cooking","Renewable-electricity-generating-capacity-per-capita","Financial flows to developing countries (US $)","Land Area(Km2)","Latitude","Longitude"]
df_Germany.drop(Removed_Columes,axis=1,inplace=True)
```

```python
df_Germany
```

# Cleaning the new Dataset (Germany)

```python
##cleaning the dataset 
print(df_Germany.duplicated())#checking if there any duplicate 
df_Germany=df_Germany.dropna() #droping rows with no values 
df_Germany=df_Germany.rename(columns={'Entity':'Country'})#changing the clume name from Eitiy to a Country 
df_Germany=df_Germany.set_index("Year") #changing the index from number to the years of the dataset 
 
```

```python
# seting the years as the index of the dataset
df_Germany
```

```python
df_new_gr=df_Germany[["Electricity from fossil fuels (TWh)","Electricity from nuclear (TWh)","Electricity from renewables (TWh)","Value_co2_emissions_kt_by_country"]]
```

```python
df_new_gr
```

```python
#creating the correlation
Corr_Matrix=df_new_gr.corr()
Corr_Matrix
```

```python
#creating a heatmap for the correlation 
sns.set (rc = {'figure.figsize':(5, 4)})
sns.heatmap(Corr_Matrix,annot=True)
```

```python
#drawing the plot 
plt.plot(df_Germany["Electricity from renewables (TWh)"],label="Renewabels Energy")
plt.plot(df_Germany["Electricity from nuclear (TWh)"],label="Nuclear Energy")
plt.plot(df_Germany["Electricity from fossil fuels (TWh)"],label= "Fossil Energy")

plt.legend(loc=1)
plt.title("Energy Production in Germany from 2000 to 2019")
plt.xlabel("Year")
plt.ylabel("Energy in TWh")
plt.rcParams['figure.figsize'] = [16, 6]
```

```python
plt.plot(df_Germany["Value_co2_emissions_kt_by_country"],label ="CO2 Emissions")
plt.legend(loc=1)
```
